```python
import whisper

# download model from https://github.com/akashmjn/tinydiarize/blob/main/whisper/__init__.py
model = whisper.load_model("/home/kozi/Downloads/small.en-tdrz.pt") # small.en-tdrz
result = whisper.transcribe(model, '/home/kozi/Documents/_onlab_git/ami/ami_dataset/test/audio/Array1-01/EN2002a.Array1-01.wav', verbose=True, condition_on_previous_text=False)
```

```python
result
```

```python
# dump the diarization output to disk using RTTM format
with open("/home/kozi/Documents/_onlab_git/ami/tinydiarize.rttm", "w") as rttm:
    (rttm)
```

```python
# Determine the speaker based on turns
def determine_speaker(idx):
    return idx % 2  # Assuming two speakers indexed as 0 and 1

# Create RTTM lines
rttm_lines = []
txt_lines = []
for idx, segment in enumerate(result['segments']):
    speaker = determine_speaker(idx)
    start = segment['start']
    end = segment['end']
    duration = end - start
    text = segment['text'].replace(' [SPEAKER TURN]', '').strip()

    # RTTM line format
    rttm_line = f"SPEAKER EN2002a 1 {start:.3f} {duration:.3f} <NA> <NA> speaker{speaker} <NA> <NA>"
    rttm_lines.append(rttm_line)

    # TXT line format
    txt_line = f"Speaker {speaker}: {text}"
    txt_lines.append(txt_line)

# Write RTTM file
with open('/home/kozi/Documents/_onlab_git/ami/tinydiarize_output.rttm', 'w') as rttm_file:
    for line in rttm_lines:
        rttm_file.write(line + '\n')
```
